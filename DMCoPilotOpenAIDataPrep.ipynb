```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
import openai


# making a get request

urls = ['https://learn.microsoft.com/en-us/microsoftteams/rooms/rooms-plan',
        'https://learn.microsoft.com/en-us/microsoftteams/rooms/rooms-licensing',
        'https://learn.microsoft.com/en-us/microsoftteams/rooms/admin-center-license-overview',
        'https://learn.microsoft.com/en-us/microsoftteams/rooms/rooms-legacy-licensing',
        'https://learn.microsoft.com/en-us/microsoftteams/rooms/teams-devices-feature-comparison',
        'https://learn.microsoft.com/en-us/microsoftteams/rooms/room-planning-guidance?tabs=uses',
        'https://learn.microsoft.com/en-us/microsoftteams/rooms/room-planning-guidance?tabs=uses',
        'https://learn.microsoft.com/en-us/microsoftteams/rooms/rooms-release-note']

pages = ['Overview of Microsoft Teams Rooms planning',
         'Overview of Teams Rooms licensing',
         'Teams Rooms admin center licensing',
         'Teams Rooms legacy licensing',
         'Teams devices feature comparison between Windows and Android',
         'Meeting room guidance for Teams',
         'Version support',
         'Release notes']

final_data = pd.DataFrame(columns = ['Page', 'Section', 'Text', 'WordCount'])


for i in range(len(urls)):
    r = requests.get(urls[i])
    soup = BeautifulSoup(r.content, 'html.parser')
    Header_1 = soup.find_all('h1')
    Header_2 = soup.find_all('h2')

    overview_content = ""

    for c in Header_1[0].nextSiblingGenerator():
        if c.name == 'h2':
            wc = len(overview_content.strip().split(" "))
            row = [pages[i], c.text, overview_content, wc]
            final_data.loc[len(final_data)] = row
            overview_content = ""
            break
        if c.name == 'p':
            overview_content += c.text
            
    content = ""

    for h in Header_2:
        for c in h.nextSiblingGenerator():
            if c.name == 'h2':
                wc = len(content.strip().split(" "))
                row = [pages[i], h.text, content, wc]
                final_data.loc[len(final_data)] = row
                content = ""
                break
            if c.name == 'p':
                content += c.text

final_data['context'] = final_data.Page + "\n" + final_data.Section + "\n\n" + final_data.Text

openai.organization = ""
openai.api_key = ""

""" def get_questions(context):
    try:
        response = openai.Completion.create(
            engine="davinci-instruct-beta-v3",
            prompt=f"Write questions based on the text below\n\nText: {context}\n\nQuestions:\n1.",
            temperature=0,
            max_tokens=257,
            top_p=1,
            frequency_penalty=0,
            presence_penalty=0,
            stop=["\n\n"]
        )
        return response['choices'][0]['text']
    except:
        return ""


final_data['questions']= final_data.context.apply(get_questions)
final_data['questions'] = "1." + final_data.questions
print(final_data[['questions']].values[0][0])

#print(final_data)
final_data.to_csv('temp.csv') """

""" questions_data = pd.read_csv('./questions.csv')

def get_answers(row):
    try:
        response = openai.Completion.create(
            engine="davinci-instruct-beta-v3",
            prompt=f"Write answer based on the text below\n\nText: {row.context}\n\nQuestions:\n{row.questions}\n\nAnswers:\n1.",
            temperature=0,
            max_tokens=257,
            top_p=1,
            frequency_penalty=0,
            presence_penalty=0
        )
        return response['choices'][0]['text']
    except Exception as e:
        print (e)
        return ""


questions_data['answers']= questions_data.apply(get_answers, axis=1)
questions_data['answers'] = "1." + questions_data.answers
questions_data = questions_data.dropna().reset_index().drop('index',axis=1)
print(questions_data[['answers']].values[0][0])
questions_data.to_csv('answers.csv') """







```
